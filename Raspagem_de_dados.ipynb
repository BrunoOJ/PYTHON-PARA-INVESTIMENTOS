```python
maior_DY["ranking_maior_DY"] = range(1, len(maior_DY)+1)
maior_VP["ranking_maior_VP"]= range(1, len(maior_DY)+1)
maior_L["ranking_maior_L"]= range(1, len(maior_DY)+1)
maior_PVP["ranking_maior_PVP"]= range(1, len(maior_DY)+1)
maior_V["ranking_maior_V"]= range(1, len(maior_DY)+1)

top_maior_DY = maior_DY.head(10)
top_maior_VP = maior_VP.head(10)
top_maior_L = maior_L.head(10)
top_maior_PVP = maior_PVP.head(10)
top_maior_V = maior_V.head(10)

top_empresas = pd.concat([top_maior_DY,top_maior_VP,top_maior_L,top_maior_PVP,top_maior_V])
top_empresas = top_empresas.drop_duplicates(subset="Unnamed: 0")
colunas = ["ranking_maior_DY","ranking_maior_VP","ranking_maior_L","ranking_maior_PVP","ranking_maior_V"]
top_empresas["ranking_total"] = top_empresas[colunas].sum(axis=1)
top_empresas = top_empresas.sort_values(by="ranking_total").head(10)
top_empresas
```
