```python
# IMPORTANDO AS BIBLIOTECAS
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import timedelta, datetime
```

```python
data_atual = datetime.now()
data_periodo = data_atual - timedelta(days=3000)
```

```python
tkr = ["GGBR4.SA","WEGE3.SA","PETR4.SA","AMER3.SA","MGLU3.SA"]

ticker = tkr[0]
data_inicio = data_periodo
data_fim = data_atual
serie_historica_preco = yf.download(ticker, start=data_inicio, end=data_fim)
serie_historica_preco
```

```python
data_compra = '2024-06-10'
df_filtro = serie_historica_preco.query('Date == @data_compra')

data_compra = df_filtro.index[0]
valor_compra = df_filtro["Close"].iloc[0]
data_atual = serie_historica_preco.index[-1]
valor_atual = serie_historica_preco["Close"].iloc[-1]
```

```python
porc = ((valor_atual-valor_compra)/valor_compra)*100


plt.figure(figsize=(12, 6))
plt.plot(serie_historica_preco['Close'], label='PreÃ§o de Fechamento')

plt.scatter(data_compra, valor_compra, color="black")
plt.text(data_compra, valor_compra, str(valor_compra)[:5], fontsize=12)
plt.axhline(valor_compra, color="blue", linestyle='--', label=f'Compra: {valor_compra:.2f}')

plt.scatter(data_atual, valor_atual, color="red")
plt.text(data_atual, valor_atual, str(valor_atual)[:5], fontsize=12)
plt.axhline(valor_atual, color="green", linestyle='--', label=f'Atual: {valor_atual:.2f}')

plt.title(f'PreÃ§o de Fechamento das AÃ§Ãµes da {ticker}' + f'  --->  {round(porc,2)} % ')
plt.xlabel('Data')
plt.ylabel('PreÃ§o de Fechamento')
plt.legend()
plt.grid()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def obter_dados_acao(ticker, periodo='1y'):
    dados = yf.download(ticker, period=periodo)
    return dados

def calcular_medias_moveis(dados, curto_prazo=20, longo_prazo=50):
    dados['SMA_Curto'] = dados['Close'].rolling(window=curto_prazo).mean()
    dados['SMA_Longo'] = dados['Close'].rolling(window=longo_prazo).mean()
    return dados

def identificar_sinais(dados):
    sinais = pd.DataFrame(index=dados.index)
    sinais['Preco'] = dados['Close']
    sinais['SMA_Curto'] = dados['SMA_Curto']
    sinais['SMA_Longo'] = dados['SMA_Longo']
    sinais['Sinal'] = 0
    sinais['Sinal'][curto_prazo:] = np.where(sinais['SMA_Curto'][curto_prazo:] > sinais['SMA_Longo'][curto_prazo:], 1, 0)
    sinais['Posicao'] = sinais['Sinal'].diff()
    return sinais

def plotar_sinais(dados, sinais, ticker):
    plt.figure(figsize=(14, 7))
    plt.plot(dados['Close'], label='PreÃ§o de Fechamento', alpha=0.5)
    plt.plot(dados['SMA_Curto'], label=f'SMA {curto_prazo} dias', alpha=0.75)
    plt.plot(dados['SMA_Longo'], label=f'SMA {longo_prazo} dias', alpha=0.75)

    plt.plot(sinais[sinais['Posicao'] == 1].index, sinais['SMA_Curto'][sinais['Posicao'] == 1], '^', markersize=10, color='g', lw=0, label='Compra')
    plt.plot(sinais[sinais['Posicao'] == -1].index, sinais['SMA_Curto'][sinais['Posicao'] == -1], 'v', markersize=10, color='r', lw=0, label='Venda')

    plt.title(f'AnÃ¡lise TÃ©cnica de {ticker}')
    plt.xlabel('Data')
    plt.ylabel('PreÃ§o')
    plt.legend()
    plt.show()

# ParÃ¢metros
ticker = 'GGBR4.SA'
periodo = '3y'
curto_prazo = 20
longo_prazo = 50

# Obter dados
dados = obter_dados_acao(ticker, periodo)

# Calcular mÃ©dias mÃ³veis
dados = calcular_medias_moveis(dados, curto_prazo, longo_prazo)

# Identificar sinais de compra e venda
sinais = identificar_sinais(dados)

# Plotar os sinais
plotar_sinais(dados, sinais, ticker)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ta.momentum import RSIIndicator
from ta.trend import MACD
import backtrader as bt

# FunÃ§Ã£o para obter dados histÃ³ricos de uma aÃ§Ã£o
def obter_dados_acao(ticker, periodo='1y'):
    dados = yf.download(ticker, period=periodo)
    return dados

# FunÃ§Ã£o para calcular indicadores tÃ©cnicos
def calcular_indicadores(dados):
    dados['SMA_Curto'] = dados['Close'].rolling(window=20).mean()
    dados['SMA_Longo'] = dados['Close'].rolling(window=50).mean()
    dados['RSI'] = RSIIndicator(dados['Close']).rsi()
    macd = MACD(dados['Close'])
    dados['MACD'] = macd.macd()
    dados['MACD_Sinal'] = macd.macd_signal()
    return dados

# FunÃ§Ã£o para identificar sinais de compra e venda
def identificar_sinais(dados):
    sinais = pd.DataFrame(index=dados.index)
    sinais['Preco'] = dados['Close']
    sinais['SMA_Curto'] = dados['SMA_Curto']
    sinais['SMA_Longo'] = dados['SMA_Longo']
    sinais['RSI'] = dados['RSI']
    sinais['MACD'] = dados['MACD']
    sinais['MACD_Sinal'] = dados['MACD_Sinal']

    sinais['Sinal'] = 0
    sinais['Sinal'][20:] = np.where((sinais['SMA_Curto'][20:] > sinais['SMA_Longo'][20:]) & 
                                    (sinais['RSI'][20:] < 70) & 
                                    (sinais['MACD'][20:] > sinais['MACD_Sinal'][20:]), 1, 0)
    sinais['Posicao'] = sinais['Sinal'].diff()
    return sinais

# FunÃ§Ã£o para plotar sinais
def plotar_sinais(dados, sinais, ticker):
    plt.figure(figsize=(14, 7))
    plt.plot(dados['Close'], label='PreÃ§o de Fechamento', alpha=0.5)
    plt.plot(dados['SMA_Curto'], label='SMA 20 dias', alpha=0.75)
    plt.plot(dados['SMA_Longo'], label='SMA 50 dias', alpha=0.75)

    plt.plot(sinais[sinais['Posicao'] == 1].index, sinais['SMA_Curto'][sinais['Posicao'] == 1], '^', markersize=10, color='g', lw=0, label='Compra')
    plt.plot(sinais[sinais['Posicao'] == -1].index, sinais['SMA_Curto'][sinais['Posicao'] == -1], 'v', markersize=10, color='r', lw=0, label='Venda')

    plt.title(f'AnÃ¡lise TÃ©cnica de {ticker}')
    plt.xlabel('Data')
    plt.ylabel('PreÃ§o')
    plt.legend()
    plt.show()

# FunÃ§Ã£o de backtesting usando Backtrader
class Strategy(bt.Strategy):
    params = (('short_period', 20), ('long_period', 50), ('rsi_period', 14), ('rsi_overbought', 70), ('rsi_oversold', 30), 
              ('stop_loss', 0.05), ('take_profit', 0.1))

    def log(self, txt, dt=None):
        ''' MÃ©todo para logar mensagens '''
        dt = dt or self.datas[0].datetime.date(0)
        print('%s, %s' % (dt.isoformat(), txt))

    def __init__(self):
        self.sma_short = bt.indicators.SimpleMovingAverage(self.data.close, period=self.params.short_period)
        self.sma_long = bt.indicators.SimpleMovingAverage(self.data.close, period=self.params.long_period)
        self.rsi = bt.indicators.RelativeStrengthIndex(self.data.close, period=self.params.rsi_period)
        self.macd = bt.indicators.MACD(self.data.close)
        self.macd_signal = self.macd.signal

    def next(self):
        if self.sma_short > self.sma_long and self.rsi < self.params.rsi_overbought and self.macd > self.macd_signal:
            if not self.position:
                self.buy()
        elif self.sma_short < self.sma_long or self.rsi > self.params.rsi_overbought:
            if self.position:
                self.sell()

    def notify_order(self, order):
        if order.status in [order.Completed]:
            if order.isbuy():
                self.log('COMPRA EXECUTADA, PreÃ§o: %.2f' % order.executed.price)
                self.buy_price = order.executed.price
                self.buy_comm = order.executed.comm
                self.stop_loss = self.buy_price * (1.0 - self.params.stop_loss)
                self.take_profit = self.buy_price * (1.0 + self.params.take_profit)
            elif order.issell():
                self.log('VENDA EXECUTADA, PreÃ§o: %.2f' % order.executed.price)
                self.sell_price = order.executed.price
                self.sell_comm = order.executed.comm

    def notify_trade(self, trade):
        if not trade.isclosed:
            return

        self.log('OPERAÃ‡ÃƒO FECHADA, Lucro Bruto %.2f, ComissÃ£o %.2f, Lucro LÃ­quido %.2f' %
                 (trade.pnl, trade.commission, trade.pnlcomm))

# ParÃ¢metros
ticker = 'AAPL'
periodo = '1y'

# Obter dados
dados = obter_dados_acao(ticker, periodo)

# Calcular indicadores
dados = calcular_indicadores(dados)

# Identificar sinais
sinais = identificar_sinais(dados)

# Plotar sinais
plotar_sinais(dados, sinais, ticker)

# Backtesting
cerebro = bt.Cerebro()
dados_bt = bt.feeds.PandasData(dataname=dados)
cerebro.adddata(dados_bt)
cerebro.addstrategy(Strategy)
cerebro.addsizer(bt.sizers.FixedSize, stake=10)
cerebro.run()
cerebro.plot()
```
