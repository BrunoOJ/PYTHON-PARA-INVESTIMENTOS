```python
# IMPORTANDO AS BIBLIOTECAS
import matplotlib.pyplot as plt
```

```python
def calcular_valorizacao(valor_inicial, porcentagem_anual, meses):
    valorizacao_mensal = (1 + porcentagem_anual / 100) ** (1 / 12) - 1
    valores = [valor_inicial]

    for i in range(1, meses + 1):
        novo_valor = valores[-1] * (1 + valorizacao_mensal)
        valores.append(novo_valor)

    return valores

def plotar_valorizacao(valores, meses):
    plt.figure(figsize=(15, 3))
    plt.bar(range(meses + 1), valores, alpha=0.6, color='b', label='Valorização')
#    plt.plot(range(meses + 1), valores, marker='o', color='r', label='Valor Médio')

    for i, valor in enumerate(valores):
        plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')

    plt.xlabel('Meses')
    plt.ylabel('Valor')
    plt.title('Valorização Mensal do Investimento')
    plt.ylim(800,2000)
    plt.legend()
    plt.show()

# Exemplo de uso
valor_inicial = 1000  # valor inicial do investimento
porcentagem_anual = 11.99  # porcentagem de valorização anual
meses = 12  # quantidade de meses

valores = calcular_valorizacao(valor_inicial, porcentagem_anual, meses)
plotar_valorizacao(valores, meses)
```

```python
def calcular_valorizacao_com_aportes(valor_inicial, porcentagem_anual, meses, aporte_mensal):
    valorizacao_mensal = (1 + porcentagem_anual / 100) ** (1 / 12) - 1
    valores = [valor_inicial]

    for i in range(1, meses + 1):
        novo_valor = valores[-1] * (1 + valorizacao_mensal) + aporte_mensal
        valores.append(novo_valor)

    return valores

def plotar_valorizacao(valores, meses):
    plt.figure(figsize=(10, 6))
    plt.bar(range(meses + 1), valores, alpha=0.6, color='b', label='Valorização')
    plt.plot(range(meses + 1), valores, marker='o', color='r', label='Valor Médio')

    for i, valor in enumerate(valores):
        plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')

    plt.xlabel('Meses')
    plt.ylabel('Valor')
    plt.title('Valorização Mensal do Investimento com Aportes Mensais')
    plt.legend()
    plt.show()

# Exemplo de uso
valor_inicial = 1000  # valor inicial do investimento
porcentagem_anual = 11.99  # porcentagem de valorização anual
meses = 60 # quantidade de meses
aporte_mensal = 500  # valor aportado mensalmente

valores = calcular_valorizacao_com_aportes(valor_inicial, porcentagem_anual, meses, aporte_mensal)
plotar_valorizacao(valores, meses)
```

```python
def calcular_valorizacao_com_aportes(valor_inicial, porcentagem_anual, meses, aporte_mensal):
    valorizacao_mensal = (1 + porcentagem_anual / 100) ** (1 / 12) - 1
    valores = [valor_inicial]
    soma_aportes = valor_inicial  # Inclui o valor inicial

    for i in range(1, meses + 1):
        novo_valor = valores[-1] * (1 + valorizacao_mensal) + aporte_mensal
        valores.append(novo_valor)
        soma_aportes += aporte_mensal

    return valores, soma_aportes

def plotar_valorizacao(valores, meses, soma_aportes):
    plt.figure(figsize=(10, 6))
    plt.bar(range(meses + 1), valores, alpha=0.6, color='b', label='Valorização')
    plt.plot(range(meses + 1), valores, marker='o', color='r', label='Valor Médio')

    for i, valor in enumerate(valores):
        plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')

    plt.axhline(y=soma_aportes, color='g', linestyle='--', label=f'Soma dos Aportes ({soma_aportes:.2f})')

    plt.xlabel('Meses')
    plt.ylabel('Valor')
    plt.title('Valorização Mensal do Investimento com Aportes Mensais')
    plt.legend()
    plt.show()

# Exemplo de uso
valor_inicial = 100000  # valor inicial do investimento
porcentagem_anual = 12  # porcentagem de valorização anual
meses = 12  # quantidade de meses
aporte_mensal = 10000  # valor aportado mensalmente

valores, soma_aportes = calcular_valorizacao_com_aportes(valor_inicial, porcentagem_anual, meses, aporte_mensal)
plotar_valorizacao(valores, meses, soma_aportes)
```

```python
def calcular_valorizacao_com_aportes(valor_inicial, porcentagem_anual, meses, aporte_mensal):
    valorizacao_mensal = (1 + porcentagem_anual / 100) ** (1 / 12) - 1
    valores = [valor_inicial]
    soma_aportes = valor_inicial  # Inclui o valor inicial

    for i in range(1, meses + 1):
        novo_valor = valores[-1] * (1 + valorizacao_mensal) + aporte_mensal
        valores.append(novo_valor)
        soma_aportes += aporte_mensal

    return valores, soma_aportes

def plotar_valorizacao(valores, meses, soma_aportes):
    plt.figure(figsize=(10, 6))
    plt.bar(range(meses + 1), valores, alpha=0.6, color='b', label='Valorização')
    plt.plot(range(meses + 1), valores, marker='o', color='r', label='Valor Médio')

    for i, valor in enumerate(valores):
        plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')

    valor_final = valores[-1]
    diferenca = valor_final - soma_aportes

    plt.axhline(y=soma_aportes, color='g', linestyle='--', label=f'Soma dos Aportes ({soma_aportes:.2f})')
    plt.axhline(y=valor_final, color='c', linestyle='--', label=f'Valor Final ({valor_final:.2f})')

    plt.xlabel('Meses')
    plt.ylabel('Valor')
    plt.title('Valorização Mensal do Investimento com Aportes Mensais')
    plt.legend(title=f'Diferença: {diferenca:.2f}')
    plt.show()

# Exemplo de uso
valor_inicial = 1000000  # valor inicial do investimento
porcentagem_anual = 11.99  # porcentagem de valorização anual
meses = 12  # quantidade de meses
aporte_mensal = 10000  # valor aportado mensalmente

valores, soma_aportes = calcular_valorizacao_com_aportes(valor_inicial, porcentagem_anual, meses, aporte_mensal)
plotar_valorizacao(valores, meses, soma_aportes)
```
